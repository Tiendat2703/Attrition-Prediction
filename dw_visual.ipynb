```python
import matplotlib.pyplot as plt
import seaborn as sns

def histogram(df, numerical_columns, ncols):
    nrows = (len(numerical_columns) + ncols - 1) // ncols  # Tính số hàng dựa trên số cột
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, nrows * 5))
    axes = axes.flatten()  # Chuyển ma trận các trục thành danh sách để duyệt dễ dàng hơn

    for i, column in enumerate(numerical_columns):
        sns.histplot(x=df[column], ax=axes[i])
        axes[i].set_title(f'histograme of {column} ')        
    # Ẩn các trục thừa
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])
        
    plt.tight_layout()
    plt.show()

# Gọi hàm với danh sách cột cần vẽ
numerical_columns = ['Age', 'DailyRate', 'DistanceFromHome', 'HourlyRate', 'MonthlyIncome',
                    'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike', 
                    'TotalWorkingYears', 'YearsAtCompany', 'YearsInCurrentRole',
                    'YearsSinceLastPromotion', 'YearsWithCurrManager']
histogram(df, numerical_columns, ncols=4)
```

```python
# Create age groups
df['AgeGroup'] = pd.cut(df['Age'], 
                       bins=[0, 25, 35, 45, 55, 100],
                       labels=['<25', '25-35', '35-45', '45-55', '>55'])

# Calculate salary brackets
df['SalaryBracket'] = pd.qcut(df['MonthlyIncome'], q=5, 
                             labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])

# Calculate experience ratio
df['ExperienceRatio'] = df['TotalWorkingYears'] / df['Age']

# Create promotion due indicator (if no promotion for 5+ years)
df['PromotionDue'] = df['YearsSinceLastPromotion'] >= 5
```

```python
plt.figure(figsize=(15, 5))

# Age Distribution
plt.subplot(131)
sns.histplot(data=df, x='Age', bins=30)
plt.title('Age Distribution')

# Gender Distribution
plt.subplot(132)
sns.countplot(data=df, x='Gender')
plt.title('Gender Distribution')

# Marital Status
plt.subplot(133)
sns.countplot(data=df, x='MaritalStatus')
plt.title('Marital Status Distribution')


plt.tight_layout()
plt.show()
```
